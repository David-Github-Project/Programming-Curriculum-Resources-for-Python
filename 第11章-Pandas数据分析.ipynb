## 11.3 Pandas数据清洗

```python
import pandas as pd
titanic=pd.DataFrame(pd.read_excel('data/titanic-2.xlsx'))
titanic.head()
```

```python
# 删除无效列
titanic.drop('embark_town', axis = 1, inplace=True)
titanic.head()
```

```python
# 查找重复值
titanic.duplicated()
```

```python
# 删除重复值
titanic = titanic.drop_duplicates()
titanic.head()
```

```python
# 统计空值的个数
titanic['who'].isnull().value_counts()
```

```python
# 使用fillna方法填充空值
titanic['who'] = titanic['who'] .fillna('man')
titanic
```

```python
titanic['age'].isnull().value_counts()
```

```python
# 使用fillna方法为age字段填充平均值
titanic['age'] = titanic['age'] .fillna(titanic['age'].mean())
titanic.head()
```

```python
#使用describe查看统计信息
titanic.describe()
```

```python
# 将异常值替换为平均值
titanic.replace([512.329200],titanic['fare'].mean())
```

## 11.4 Pandas数据分析

```python
# 导入包
import numpy as np
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
import sklearn
```

```python
# 数据加载并转换成DataFrame
boston_df=pd.read_excel('data/boston_data.xlsx', sheet_name='Sheet1')
boston_df.head()
```

```python
boston_df.describe()
```

```python
# 构建线性回归预测模型
from sklearn.linear_model import LinearRegression
X = boston_df.drop("PRICE", axis = 1)
predict_model = LinearRegression()
predict_model.fit(X, boston_df['PRICE'])    #训练模型
print("回归系数为：", predict_model.coef_)   # 判断相关性
```

```python
import seaborn as sns
sns.regplot( boston_df.RM, boston_df.PRICE)
```

```python
# 房价预测
predict_model.predict(X)[0:10]
```

```python
# 模型评估-训练集和测试集的划分
X_train, X_test, Y_train, Y_test = sklearn.model_selection.train_test_split(X, boston_df.PRICE, test_size=0.33, random_state=5)
# 构建模型
predict_model_2 = LinearRegression()
predict_model_2.fit(X_train, Y_train)
predict_train = predict_model_2.predict(X_train)
predict_test = predict_model_2.predict(X_test)
# 打印误差
print ("模型训练集的误差：", np.mean((Y_train - predict_model_2.predict(X_train)) ** 2))
print( "模型测试集的误差：", np.mean((Y_test - predict_model_2.predict(X_test)) ** 2))
```

## 11.5 Pandas绘图

```python
#例11-5
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif']=['Arial Unicode MS'] #用来正常显示中文标签
plt.rcParams['axes.unicode_minus']=False #用来正常显示负号
data=np.array([641281, 685993, 740061, 820754,900309])
index=['2014','2015','2016','2017','2018']
s = pd.Series(data, index)
s.name='2014-2018年国内生产总值'
s.plot(kind='bar',title='2014-2018年国内生产总值')
plt.show()
```

```python
#例11-6
index_list = ['国内生产总值（亿元）','全员劳动生产率(元/人)']
columns_list = ['2014','2015','2016','2017','2018']
df=pd.DataFrame([[641281,685993,740061,820754,900309], [83211.0, 88685.0, 94443.0, 100702.0, 107327.0]], index=index_list, columns=columns_list).T

df.name="2014-2018年国内生产总值和全员劳动生产率"

# 绘制“国内生产总值”柱状图
df['国内生产总值（亿元）'].plot(kind='bar', title="2014-2018年国内生产总值和全员劳动生产率")
# 绘制“比上年增长”折线图

df['全员劳动生产率(元/人)'].plot(secondary_y=True, label='全员劳动生产率(元/人)', style='c*-')
plt.legend(loc = 'upper right')
plt.show()

```
