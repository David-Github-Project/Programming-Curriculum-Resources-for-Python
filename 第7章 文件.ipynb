```python
import os
import os.path
rootdir = "c:/python"             # 指明被遍历的文件夹

for parent,dirnames,filenames in os.walk(rootdir):
# 三个参数：分别返回1.父目录 2.所有文件夹名字（不含路径） 3.所有文件名字
    for dirname in  dirnames:                       #输出文件夹信息
        print("parent is:" + parent)
        print("dirname is:" + dirname)
    
    for filename in filenames:                        #输出文件信息
        print("parent is:" + parent)
        print("filename is:" + filename)
        print("the full name of the file is:" + os.path.join(parent,filename))
#输出完整文件路径信息
```

```python
import csv
gdp={"年份":"GDP","2013":59.3,"2014":64.13,"2015":68.60,"2016":74.00,"2017":82.07,"2018":90.03}

#将字典gdp写入gdpfile.csv文件中。
with open('gdpfile.csv', 'w',encoding='utf-8',newline='') as f:
    writer = csv.writer(f) 
    for k,r in gdp.items(): 
        writer.writerow([k,r])

gdp_new = {} #新字典，用于存放从gdpfile.csv文件中读取的数据
#将gdpfile.csv文件中的数据存储到字典gdp_new中。
with open('gdpfile.csv','r',encoding='utf-8') as f: 
    reader = f.readline().strip("\n") 
    reader = f.readline().strip("\n") 
    while reader!= "": 
        tmpList =reader.split(',')
        gdp_new[tmpList[0]] = tmpList[1] 
        reader = f.readline().strip("\n")

year=input("请输入查询年份(2013-2018):")
value1=gdp_new.get(year,0) 
value2=gdp_new.get(str(int(year)-1),0) 
if value1==0: 
    print("输入的年份数据有误！") 
else: 
    if value2==0: 
        print("{}年国内生产总值为{:.2f}万亿！".format(year,float(value1))) 
    else: 
        r=(float(value1)-float(value2))/float(value2) 
        print("{}年国内生产总值为{:.2f}万亿！年度增长率为{:.2%}".format(year,float(value1),r))
```

```python
from PIL import Image, ImageFont, ImageDraw, ImageFilter
import random

def rndNum():# 随机数字
    return chr(random.randint(48, 57))

def rndColor(a,b): # 随机颜色
    return (random.randint(a,b), random.randint(a,b), random.randint(a,b))

width = 240
height = 60
p1 = Image.new('RGB', (width, height), (255, 255, 255)) # 图片尺寸240 x 60
font = ImageFont.truetype('c:/windows/fonts/Arial.ttf', 36) # 创建Font对象
draw = ImageDraw.Draw(p1) # 创建Draw对象
for x in range(width):
    for y in range(height):  
        draw.point((x, y), fill=rndColor(127, 255)) #给每个像素填充杂色

for t in range(4):  # 在图片上画上4个数字
    draw.text((60 * t + 10, 10), rndNum(), font=font, fill=rndColor(32, 127))
p1 = p1.filter(ImageFilter.BLUR)  #滤镜模糊处理
p1.save('code.jpg', 'jpeg');
p1.show()
```

```python
'''将红楼梦全集文本分词，并附带上词性，因为数据量比较大，防止每次运行脚本都花大量时间，所以第一次分词后就将结果存入文件cut_words.txt中,相当于做一个缓存，格式为每个词占一行，每一行的内容为：词  词性
'''
honglou_text = open('hongloumeng.txt','r',encoding='UTF-8').read()
# 读取文件，编码格式utf-8，防止处理中文出错

import jieba.posseg as psg
honglou_words_with_attr = [(x.word,x.flag) for x in psg.cut(honglou_text) if len(x.word) >= 2]  # x.word为词本身，x.flag为词性
print(len(honglou_words_with_attr))  # 输出：
with open('cut_words.txt','w+') as f:
    for x in honglou_words_with_attr:
        f.write('{0}\t{1}\n'.format(x[0],x[1]))
```

```python
# 从cut_words.txt中读取带词性的分词结果列表
honglou_words_with_attr = []
with open('cut_words.txt','r') as f:
    for x in f.readlines():
        pair = x.split()
        honglou_words_with_attr.append((pair[0],pair[1]))
# 将分词列表的词性构建成一个字典，以便后面使用，格式为：
# {词:词性}
attr_dict = {}
for x in honglou_words_with_attr:
    attr_dict[x[0]] = x[1]
# stop_attr中存放要过滤掉的词性列表
stop_attr =['a','ad','b','c','d','f','df','m','mq','p','r','rr','s','t','u','v','z']
#过滤清洗数据，将结果存放在words中
words = [x[0] for x in honglou_words_with_attr if x[1] not in stop_attr]
```

```python
# 从cut_words.txt中读取带词性的分词结果列表
from collections import Counter
c = Counter(words).most_common(50)
with open('result.txt','w+') as f:
    for x in c:
        f.write('{0},{1}\n'.format(x[0],x[1]))  #存为文件，方便查看和改进结果
```

```python
honglou_text = open('hongloumeng.txt','r',encoding='UTF-8').read()
import jieba.posseg as psg
honglou_words_with_attr  = [(x.word,x.flag) for x in psg.cut(honglou_text) if len(x.word) >= 2]
stop_attr = ['m','r','t','v','n','c','v','ns','q','d','u','s','nrt','f','df','a','y','z','l','b','mq']  # 要过滤掉的词性列表

delword={"老太太","小丫头","明白","言语","冷笑","和尚","老婆子","小姐","怡红院","那丫头","仔细","胡说","老嬷嬷","大老爷","从小儿","那婆子","王爷","认真"}
#过滤非人名列表。

words = [x[0] for x in honglou_words_with_attr  if x[1] not in stop_attr  and x[0] not in delword]   #数据过滤
n=0
for x in words:   #合并同义人名
    if len(x)==1:
        n=n+1
        continue
    elif x=="林姑娘" or x=="林黛玉" or x=="林妹妹" or x=="黛玉笑"or x=="黛玉忙" or x=="那黛玉" or x=="林丫头":
        x="黛玉"
    elif x=="贾政道" or x=="贾政听"or x=="贾政又"or x=="贾政笑"or x=="贾政便" or x=="贾政回"or x=="贾政因":
        x="贾政"
    elif x=="宝二爷"or x=="宝兄弟" or x=="那宝玉" or x=="向宝玉" or x=="问宝玉" or x=="贾宝玉":
        x="宝玉"
    elif x=="贾母笑" or x=="老祖宗"or x=="贾母因"or x=="贾母王" or x=="贾母忙" or x=="贾母命"or x=="贾母问":
        x="贾母"
    elif x=="宝钗笑"or x=="宝姐姐" or x=="宝丫头" or x=="薛宝钗"or x=="宝钗见"or x=="宝钗忙" or x=="宝钗因":
        x="宝钗"
    elif x=="凤姐"or x=="凤姐儿" or x=="凤丫头"or x=="凤姐姐":
        x="王熙凤"
    else:
        x=x
    words[n]=x    #同义人名统一名称
    n=n+1
from collections import Counter
c=Counter(words).most_common(10)  #计算出场次数前10名的人物并输出
for x in c:
     print("{0:<10}{1:>5}".format(x[0],x[1]))
```
