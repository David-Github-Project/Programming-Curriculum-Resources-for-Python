```python
import matplotlib.pyplot as plt
import numpy as np
# 创建figure对象
plt.figure()
# 绘制普通图像
x = np.linspace(-2*np.pi, 2*np.pi, 1000)
y = np.sin(x)
plt.plot(x, y, label='y=sin(x)')
# 添加标题，并设置标题为大号、加粗、蓝色字体
plt.title('Figure of y = sin(x)',fontsize='large', fontweight='bold', color='blue')
# 为sin(0)=0添加标注
plt.annotate('sin(0)=0', xy=(0,0),xytext=(1,-0.5), arrowprops=dict(facecolor='black',shrink=0.05))
# 添加x轴标签和y轴标签
plt.xlabel('Values of x')
plt.ylabel('Values of y')
# 添加图例，位置设置在左上角
plt.legend(loc='upper left')
# 添加网格
plt.grid()
# 设置x轴和y轴的范围，其中x轴的范围是[-6,6]，y轴的范围是[-1.0,1.0]
plt.axis((-6, 6, -2.0, 2.0))
plt.show()
```

```python
plt.figure(figsize=(10,6))
# 招生年份
years = ['2014', '2015', '2016', '2017', '2018']
# 2014-2018年普通本专科、中等职业教育及普通高中招生人数（万人）
undergraduate_enrollment = [721, 738, 749, 761, 791]
secondary_vocational_enrollment = [620, 601,593,582,557]
high_school_enrollment = [797, 797, 803, 800,793]
x = list(range(len(undergraduate_enrollment)))

#设置间距
total_width, n = 0.8, 3
width = total_width / n
#在偏移间距位置绘制柱状图1
for i in range(len(x)):
    x[i] -= width
plt.bar(x, undergraduate_enrollment, width=width, label='普通本专科', fc='teal')
# 设置数字标签
for a, b in zip(x, undergraduate_enrollment):
    plt.text(a, b, b, ha='center', va='bottom', fontsize=10)
             
#在偏移间距位置绘制柱状图2
for i in range(len(x)):
    x[i] += width
plt.bar(x, secondary_vocational_enrollment, width=width, label='中等职业教育', tick_label=years, fc='darkorange')
# 设置数字标签
for a, b in zip(x, secondary_vocational_enrollment):
    plt.text(a, b, b, ha='center', va='bottom', fontsize=10)
    
#在偏移间距位置绘制柱状图3
for i in range(len(x)):
    x[i] += width
plt.bar(x, high_school_enrollment, width=width, label='普通高中', fc='lightcoral')
# 设置数字标签
for a, b in zip(x, high_school_enrollment):
    plt.text(a, b, b, ha='center', va='bottom', fontsize=10)
    
plt.title("2014-2018年普通本专科、中等职业教育及普通高中招生人数（万人）")
plt.ylabel("招生人数（万人）")
plt.legend(loc='lower right')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
# 创建3d图形的两种方式
# ax = Axes3D(fig)
ax = fig.add_subplot(111, projection='3d')
# X, Y value
X = np.arange(-4, 4, 0.25)
Y = np.arange(-4, 4, 0.25)
X, Y = np.meshgrid(X, Y)    # x-y 平面的网格
R = np.sqrt(X ** 2 + Y ** 2)
# height value
Z = np.sin(R)

# rstride:行之间的跨度  cstride:列之间的跨度
# rcount:设置间隔个数，默认50个，ccount:列的间隔个数  不能与上面两个参数同时出现
#vmax和vmin  颜色的最大值和最小值
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=plt.get_cmap('rainbow'))
# zdir : 'z' | 'x' | 'y' 表示把等高线图投射到哪个面
# offset : 表示等高线图投射到指定页面的某个刻度
ax.contourf(X,Y,Z,zdir='z',offset=-2)
# 设置图像z轴的显示范围，x、y轴设置方式相同
ax.set_zlim(-2,2)
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = Axes3D(fig)

# 构造函数
x = np.arange(-10, 10, 0.1)
y = np.arange(-10, 10, 0.1)
X, Y = np.meshgrid(x, y)
Z = np.add(-np.power(X, 2), np.power(Y, 2))
#使用plot_wireframe方法绘制网格
surf = ax.plot_wireframe(X, Y, Z, alpha=0.2)
ax.contour(X, Y, Z, cmap=cm.Accent, linewidths=2)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = Axes3D(fig)

x = np.arange(-15, 15, 0.1)
y = np.arange(-15, 15, 0.1)
X, Y = np.meshgrid(x, y)
Z = np.add(-np.power(X, 4), np.power(Y, 4))
# 绘制三维曲面图
surf = ax.plot_surface(X, Y, Z, cmap=cm.gist_rainbow)
# 设置颜色条
fig.colorbar(surf, shrink=0.5, aspect=5)
plt.show()
```
