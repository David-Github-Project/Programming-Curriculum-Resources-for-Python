# Seaborn绘图

```python
import seaborn as sns
import matplotlib.pyplot as plt 

titanic=sns.load_dataset('titanic')
titanic.sample(5)
```

```python
sns.set(style="white")
sns.jointplot(x="age",y='fare',data = titanic)
```

```python
sns.jointplot(x="age",y='fare',data = titanic, kind='reg')
```

```python
sns.jointplot(x="age",y='fare',data = titanic, kind='hex')
```

```python
sns.jointplot(x="age",y='fare',data = titanic, kind='kde', color='r')
```

```python
sns.boxplot(x='who',y='age', data=titanic)
```

```python
# 根据两个离散变量嵌套分组的盒图，hue用于指定要嵌套的离散变量
sns.boxplot(x='who',y='age', hue='alive', data=titanic)
```

```python
sns.distplot( titanic['fare'])
```

```python
fig,axes=plt.subplots(2,2)
# 1.默认绘图效果
sns.distplot( titanic['fare'], ax=axes[0][0])
#2. kde=False:不显示密度曲线
sns.distplot( titanic['fare'], kde=False, ax=axes[0][1])
#3.  rug=True:在坐标轴上添加地毯图
sns.distplot( titanic['fare'], rug=True, ax=axes[1][0])
#4.通过调节vertical，hist_kws和kde_kws参数，改变直方图的方向、坐标轴和密度曲线颜色
sns.distplot( titanic['fare'], vertical=True, hist_kws={'color':'black','label':'hist'}, kde_kws={'color':'red','label':'KDE'}, ax=axes[1][1])
```

```python
#绘制柱状图
sns.countplot(titanic['sex'])
```

```python
fig,axes=plt.subplots(1,2)
#1.使用两个变量作图（下图左子图）
sns.countplot(x=titanic['sex'], hue=titanic['class'], ax=axes[0])
#2.绘制水平图（下图右子图）
sns.countplot(y=titanic['sex'], hue=titanic['class'], ax=axes[1])
```

```python
# 绘制单核密度图
sns.kdeplot(titanic['fare'])
```

```python
fig,axes=plt.subplots(1,2)
# 绘制单核密度图(下图左子图）
sns.kdeplot(titanic['fare'], ax=axes[0])
# 绘制单核密度图，并修改颜色（下图右子图）
sns.kdeplot(titanic['fare'],shade=True,color='r', ax=axes[1])
```

```python
# 小提琴图
sns. violinplot (titanic['age'])
```

```python
# 基于类别变量进行分组的多个垂直小提琴图
sns.violinplot(x=titanic['pclass'], y=titanic['age'])
```

```python
fig,axes=plt.subplots(2,2)
# 1.默认绘图效果
sns.violinplot(x=titanic['pclass'], y=titanic['age'], ax=axes[0][0])
#2. hue参数可以对字段进行细分
sns.violinplot(x=titanic['pclass'], y=titanic['age'], hue=titanic['sex'], ax=axes[0][1])
#3. split参数可以“拆分”小提琴
sns.violinplot(x=titanic['pclass'], y=titanic['age'], hue=titanic['sex'], split=True, ax=axes[1][0])
#4. inner参数可以在小提琴内部添加图形
sns.violinplot(x=titanic['pclass'], y=titanic['age'], hue=titanic['sex'], split=True, inner='stick', ax=axes[1][1])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt 

tips=sns.load_dataset('tips')
tips.sample(10)
```

```python
sns.regplot(x='total_bill',y='tip', data=tips)
```

```python
fig,axes=plt.subplots(2,2)
# 1.默认绘图效果
sns.regplot(x='total_bill',y='tip', data=tips, ax=axes[0][0])
#2. ci参数可以控制是否显示置信区间
sns.regplot(x='total_bill',y='tip', data=tips, ci=None, ax=axes[0][1])
#3. mark参数可以设置数据点格式，color参数可以设置颜色
sns.regplot(x='total_bill',y='tip', data=tips, color='g',marker='*', ax=axes[1][0])
#4. fit_reg参数可以控制是否显示拟合的直线
sns.regplot(x='total_bill',y='tip', data=tips, color='r',marker='+', fit_reg=False, ax=axes[1][1])
```

```python
# lmpot()函数中的hue参数可以进行分类拟合
sns.lmplot(x='total_bill',y='tip', hue='time', data=tips )
```

```python
# lmpot()函数中的col参数可以实现不同类别分开绘制
sns.set_style("white")
sns.lmplot(x='total_bill',y='tip', col='time', data=tips )
```
