# Pandas绘图

```python
#例10-5
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif']=['Arial Unicode MS'] #用来正常显示中文标签
plt.rcParams['axes.unicode_minus']=False #用来正常显示负号
data=np.array([641281, 685993, 740061, 820754,900309])
index=['2014','2015','2016','2017','2018']
s = pd.Series(data, index)
s.name='2014-2018年国内生产总值'
s.plot(kind='bar',title='2014-2018年国内生产总值')
plt.show()
```

```python
#例10-6
index_list = ['国内生产总值（亿元）','全员劳动生产率(元/人)']
columns_list = ['2014','2015','2016','2017','2018']
df=pd.DataFrame([[641281,685993,740061,820754,900309], [83211.0, 88685.0, 94443.0, 100702.0, 107327.0]], index=index_list, columns=columns_list).T

df.name="2014-2018年国内生产总值和全员劳动生产率"

# 绘制“国内生产总值”柱状图
df['国内生产总值（亿元）'].plot(kind='bar', title="2014-2018年国内生产总值和全员劳动生产率")
# 绘制“比上年增长”折线图

df['全员劳动生产率(元/人)'].plot(secondary_y=True, label='全员劳动生产率(元/人)', style='c*-')
plt.legend(loc = 'upper right')
plt.show()
```
