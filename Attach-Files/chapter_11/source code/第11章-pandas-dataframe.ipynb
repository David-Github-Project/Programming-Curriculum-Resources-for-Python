```python
#DataFrame的创建-自定义index和columns
index_list = ['国内生产总值（亿元）','比上年增长(百分比)']
columns_list = ['2014年','2015年','2016年','2017年','2018年']
df = pd.DataFrame([[641281, 685993, 740061, 820754,900309], [7.3,6.9,6.7,6.8,6.6]], index=index_list, columns=columns_list)
df.name="2014-2018年国内生产总值和比上年增长"
df
```

```python
#DataFrame的创建-用对象为列表的字典
index_list = ['国内生产总值（亿元）','比上年增长(百分比)']

data = {'2014年':[641281,7.3],
        '2015年':[685993,6.9],
        '2016年':[740061,6.7],
        '2017年':[820754,6.8],
        '2018年':[900309,6.6]
       }
df = pd.DataFrame(data, index=index_list)
df.name="2014-2018年国内生产总值和比上年增长"
df
```

```python
index_list = ['国内生产总值（亿元）','比上年增长(百分比)','能耗降低率','全员劳动生产率(元/人)']
columns_list = ['2014年','2015年','2016年','2017年','2018年']
df=pd.DataFrame([[641281,685993,740061,820754,900309], [7.3,6.9,6.7,6.8,6.6],[-4.8,-5.6,-5.0,-3.7,-3.1],[83211, 88685,94443,100702,107327]], index=index_list, columns=columns_list)
df.name="2014-2018年国内生产总值数据"
df
```

```python
#获取”2018年”列的数据
df.loc[: , '2018年' ]       #方法1:使用标签获取
df.iloc[:,4]            # 方法2:使用位置获取
```

```python
#获取“2016年”到”2018年”3列的数据
df.loc[: , '2016年':'2018年' ]       #方法1:使用标签获取
df.iloc[:,2:5]       # 方法2:使用位置获取
```

```python
#获取”能耗降低率”行的数据
df.loc['能耗降低率', : ]         # 方法1:使用标签获取
df.iloc[2,:]                      # 方法2:使用位置获取
```

```python
#获取“国内生产总值（亿元）”行到”能耗降低率”行的数据
df.loc['国内生产总值（亿元）' : '能耗降低率', : ]         # 方法1:使用标签获取
df.iloc[0:3, : ]  
```

```python
#获取“国内生产总值（亿元）”行～”能耗降低率”行和“2016年”列～”2018年”列之间的数据
df.loc['国内生产总值（亿元）' : '能耗降低率', '2016年':'2018年' ] 
df.iloc[0:3, 2:5]  
```
